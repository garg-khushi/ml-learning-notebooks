```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans 
from sklearn.datasets import make_blobs 
from sklearn.preprocessing import StandardScaler
import plotly.express as px

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
#creatinf own dataset
#need to set a random seed=0
np.random.seed(0)
```

```python
#make blobs
X,y=make_blobs(n_samples=5000,centers=[[4,4],[2,-3],[-2,-1],[1,1]],cluster_std=0.9)
```

```python
plt.scatter(X[:,0], X[:,1], marker='.',alpha=0.3,ec='k',s=80)
plt.show()
```

```python
k_means=KMeans(init="k-means++", n_clusters=4,n_init=12)
              #fitting
k_means.fit(X)
```

```python
#label for each point in the model
k_means_labels=k_means.labels_
k_means_labels
```

```python
k_means_cluster_centers = k_means.cluster_centers_
k_means_cluster_centers
```

```python
#visualisation
fig = plt.figure(figsize=(6, 4))
colors = plt.cm.tab10(np.linspace(0, 1, len(set(k_means_labels))))
ax = fig.add_subplot(1, 1, 1)
for k, col in zip(range(len([[4,4],[-2,-1],[2,-3],[1,1]])),colors):
    my_members = (k_means_labels == k)
    cluster_center = k_means_cluster_centers[k]
    ax.plot(X[my_members, 0], X[my_members, 1], 'w', markerfacecolor=col, marker='.',ms=10)
    ax.plot(cluster_center[0], cluster_center[1], 'o', markerfacecolor=col,  markeredgecolor='k', markersize=6)

ax.set_title('KMeans')
ax.set_xticks(())
plt.show()
```

```python
k_means3 = KMeans(init="k-means++", n_clusters=3, n_init=12)
k_means3.fit(X)
fig = plt.figure(figsize=(6, 4))
colors = plt.cm.tab10(np.linspace(0, 1, len(set(k_means3.labels_))))
ax = fig.add_subplot(1, 1, 1)
for k, col in zip(range(len(k_means3.cluster_centers_)), colors):
    my_members = (k_means3.labels_ == k)
    cluster_center = k_means3.cluster_centers_[k]
    ax.plot(X[my_members, 0], X[my_members, 1], 'w', markerfacecolor=col, marker='.',ms=10)
    ax.plot(cluster_center[0], cluster_center[1], 'o', markerfacecolor=col,  markeredgecolor='k', markersize=6)
plt.show()
```
